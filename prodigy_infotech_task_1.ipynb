```python
#Filter the data for a specific country
chousen_country = "United States"  # 👈 Change this to the country you want
country_data = df[df['Country Name'] == chousen_country]

#  Extract only columns that are year numbers (like "1960", "1961", etc.)
year_columns = [col for col in country_data.columns if col.isdigit()]
years = list(map(int, year_columns))  # convert to integers

# Get population values for the country and flatten the array
population = country_data[year_columns].values.flatten()

# Plot as a bar chart
plt.figure(figsize=(10, 4))
plt.bar(years, population, color='skyblue')
plt.title(f"Population of {chousen_country} Over Time (Bar Chart)")
plt.xlabel("Year")
plt.ylabel("Population")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
